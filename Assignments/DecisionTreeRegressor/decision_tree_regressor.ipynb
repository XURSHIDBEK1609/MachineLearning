```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor, plot_tree
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Simple dataset: House Size vs. Price
X = [[500], [1000], [1500], [2000], [2500], [3000], [3500], [4000], [4500]]
Y = [75000, 120000, 180000, 250000, 320000, 400000, 490000, 590000, 700000]  # Target variable (Price)

# Convert to NumPy arrays
X = np.array(X)
Y = np.array(Y)
```

```python
# Splitting into training (80%) and testing (20%)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
# Training the Decision Tree Regressor
model = DecisionTreeRegressor(max_depth=3)
model.fit(X_train, Y_train)

# Predictions
Y_pred = model.predict(X_test)

# Evaluating the model
mse = mean_squared_error(Y_test, Y_pred)
r2 = r2_score(Y_test, Y_pred)

mse, r2
```

```python
# Visualizing the Decision Tree
plt.figure(figsize=(10,5))
plot_tree(model, filled=True, feature_names=['House Size'], rounded=True)
plt.show()
```
